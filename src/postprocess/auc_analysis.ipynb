```python
from sklearn.metrics import roc_auc_score, accuracy_score, roc_curve
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn')
plt.rcParams['axes.grid'] = False
%matplotlib inline
```

```python
total = 60000
ratio = 0.15
a = 0.15*np.random.randn(int(total*ratio)) + 0.46
b = 0.06*np.random.randn(int(total*(1-ratio))) + 0.20
b_zeros = np.zeros(len(b))
a_ones = np.ones(len(a))
samples = np.hstack([a, b])
true = np.hstack([a_ones, b_zeros])
```

```python
auc = roc_auc_score(true, samples)
auc
```

```python
acc = accuracy_score(true, samples > 0.5)
acc
```

```python
fpr, recall, threshold = roc_curve(true, samples)

# Plot ROC curves
plt.subplots(1, figsize=(10,10))
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, recall)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
